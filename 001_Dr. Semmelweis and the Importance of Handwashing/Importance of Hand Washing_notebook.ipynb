![Dr. Semmelweis](ignaz_semmelweis_1860.jpeg)

Hungarian physician Dr. Ignaz Semmelweis worked at the Vienna General Hospital with childbed fever patients. Childbed fever is a deadly disease affecting women who have just given birth, and in the early 1840s, as many as 10% of the women giving birth died from it at the Vienna General Hospital. Dr.Semmelweis discovered that it was the contaminated hands of the doctors delivering the babies, and on June 1st, 1847, he decreed that everyone should wash their hands, an unorthodox and controversial request; nobody in Vienna knew about bacteria.

You will reanalyze the data that made Semmelweis discover the importance of handwashing and its impact on the hospital.

The data is stored as two CSV files within the `data` folder.

`yearly_deaths_by_clinic.csv` contains the number of women giving birth at the two clinics at the Vienna General Hospital between the years 1841 and 1846.

| Column | Description |
|--------|-------------|
|`year`  |Years (1841-1846)|
|`births`|Number of births|
|`deaths`|Number of deaths|
|`clinic`|Clinic 1 or clinic 2|

`monthly_deaths.csv` contains data from 'Clinic 1' of the hospital where most deaths occurred.

| Column | Description |
|--------|-------------|
|`date`|Date (YYYY-MM-DD)
|`births`|Number of births|
|`deaths`|Number of deaths|

```r
# Imported libraries
library(tidyverse)
```

### 1. Load the CSV files into yearly and monthly data frames and check the data.

- Load Datasets

```r
# Load the yearly dataset

yearly <- read_csv('data/yearly_deaths_by_clinic.csv', show_col_types = FALSE)
yearly
```

```r
# Load the monthly dataset

monthly <- read_csv('data/monthly_deaths.csv', show_col_types = FALSE)
monthly
```

### Inspection

After careful inspection of the data, no empty row.

### 2. Add a new column with the proportions

- Add proportion_deaths to both monthly and yearly data 

```r
yearly <- yearly %>%
	mutate(proportion_deaths = deaths / births)

yearly
```

```r
monthly <- monthly %>%
	mutate(proportion_deaths = deaths / births)

monthly
```

### 3. Make a Line plot for both data frames

```r
# Line plot for yearly_data

ggplot(yearly, aes(x = year, y = proportion_deaths, color = clinic)) + geom_line() + labs(x = "Year of birth and death", y = "Proportion Deaths to Births Record")
```

_It is seen from the plot that more death that occured at Clinic 1 than Clinic 2_

```r
# Line plot for monthly_data

ggplot(monthly, aes(x = date, y = proportion_deaths)) + geom_line() + labs(x = "Date of birth and death", y = "Proportion Deaths to Births Record")
```

_it is seen from the visualazation that the date rate reduces from between 1847 and 1848._

### 4. Visualize the Threshold

```r
# set the threshold
handwashing_start = as.Date('1847-06-01')
```

```r
# Add threshold to data frame

monthly <- monthly %>%
	mutate(handwashing_started = date >= handwashing_start)
monthly
```

```r
# Plot the monthly datagain using the threshold

ggplot(monthly, aes(x = date, y = proportion_deaths, color = handwashing_started)) + geom_line() + labs(x = "Date of birth and death", y = "Proportion Deaths to Births Record")
```

_it was noticed that after handwashing started the death rate reduced_

### 5. Calculate the mean proportion of deaths

```r
# Calculate the mean proportion of deaths before and after handwashing from the monthly data

monthly_summary <- monthly %>%
	group_by(handwashing_started) %>%
	summarize(proportion_deaths_mean = mean(proportion_deaths))

monthly_summary
```

```r
percentage_decrease_in_death <- ((0.105 - 0.0211) / 0.105) * 100

cat("After handwashing started, there was", round(percentage_decrease_in_death), "% decrease in death\n")
```
